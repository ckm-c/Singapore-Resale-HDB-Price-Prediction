```python
df = df.drop(['month','block','street_name','flat_model','lease_commence_date'],axis='columns')
df.shape
df.head(10)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['remaining_lease'] = df['remaining_lease'].apply(lambda x: int(x[:2]))
```

```python
df.head(5)
```

```python
df.info()
```

```python
values = ['1 ROOM','MULTI-GENERATION']
df =df[df['flat_type'].isin(values)==False]
```

```python
df['flat_type'].value_counts()
```

```python
df['flat_type']= df['flat_type'].map({'2 ROOM': 0, '3 ROOM': 1,'4 ROOM': 2, '5 ROOM': 3, 
                                                'EXECUTIVE': 4}).astype(int)
```

```python
df = df.sort_values(by='storey_range')
df['storey_range'] = df['storey_range'].astype('category').cat.codes
```

```python
df.info()
```

```python
dummies = pd.get_dummies(df.town)
```

```python
dummies.head(5)
```

```python
df.head(5)
```

```python
df = pd.concat([df,dummies],axis='columns')
df.head(10)
```

```python
df=df.drop('town',axis='columns')
df.head(10)
```

```python
df.info()
```

```python
X = df.drop(['resale_price'],axis='columns')
X.shape
```

```python
X.hist(figsize=(20,20))
plt.show()
```

```python
y = df["resale_price"]
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
rfr.fit(X, y) 
rfr.score(X_test,y_test)
```

```python
from sklearn.linear_model import Lasso
lasso = Lasso(tol=0.01)
lasso.fit(X_train,y_train)
lasso.score(X_test,y_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
dtr.fit(X_train,y_train)
dtr.score(X_test,y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=5, test_size=0.3, random_state=42)

lasso_cv = cross_val_score(Lasso(tol=0.01), X, y, cv=cv)
print(lasso_cv)


rfr_cv = cross_val_score(RandomForestRegressor(), X, y, cv=cv)
print(rfr_cv)

dtr_cv = cross_val_score(DecisionTreeRegressor(), X, y, cv=cv)
print(dtr_cv)
```

```python
from sklearn.model_selection import GridSearchCV

def find_best_model_using_gridsearchcv(X,y):
    algos = {
         'lasso': {
            'model': Lasso(),
            'params': {
                'tol':[0.001,0.1,1],
                'alpha': [1,2],
                'selection': ['random', 'cyclic']
            }
        },
        'random_forest': {
            'model': RandomForestRegressor(),
            'params': {
                "n_estimators"      : [10,20,30],
                "max_features"      : ["auto", "sqrt", "log2"],
                "min_samples_split" : [2,4,8],
                "bootstrap": [True, False],
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion' : ['mse','friedman_mse'],
                'splitter': ['best','random']
            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.3, random_state=42)
    for algo_name, config in algos.items():
        gs =  GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(X,y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(scores,columns=['model','best_score','best_params'])

find_best_model_using_gridsearchcv(X,y)
```
